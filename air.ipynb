```python
df.dropna(axis=1, how='all', inplace=True)
```

```python
df['Period'] = pd.to_datetime(df['Period'], format='%m.%Y')
```

```python
df= df.sort_values(by=["Period"])
```

```python
df[df['StationName']=='Пролетарский проспект']
```

```python
df['MonthlyAverage'] = df['MonthlyAverage'].astype(float)
df['MonthlyAveragePDKss'] = df['MonthlyAveragePDKss'].astype(float)
df['MonthlyAverage']
```

```python
import pandas as pd


# Агрегация данных по периоду, округу и параметру
df_aggregated = df.groupby(["Period", "AdmArea", "Parameter"]).agg({
    "MonthlyAverage": "mean",
    "MonthlyAveragePDKss": "mean"
}).reset_index()  # Сброс индексов и перевод полученного DataFrame в обычный формат
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots


# Функция для рисования графика изменения параметра по округам
def draw_graph(adm_area):
    # Получение списка доступных параметров для данного округа
    available_parameters = sorted(df_aggregated[df_aggregated["AdmArea"] == adm_area]["Parameter"].unique())

    # Создание фигуры
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    buttons = []
    visible = [True] * (len(available_parameters) * 2)
    for i, parameter in enumerate(available_parameters):
        # Фильтрация данных по округу и параметру
        df_filtered = df_aggregated[(df_aggregated["AdmArea"] == adm_area) & (df_aggregated["Parameter"] == parameter)]

        # Добавление графика изменения MonthlyAverage
        fig.add_trace(
            go.Scatter(x=df_filtered["Period"], y=df_filtered["MonthlyAverage"], name="{} - MonthlyAverage".format(parameter)),
            secondary_y=True
        )

        # Добавление графика изменения MonthlyAveragePDKss
        fig.add_trace(
            go.Scatter(x=df_filtered["Period"], y=df_filtered["MonthlyAveragePDKss"], name="{} - MonthlyAveragePDKss".format(parameter)),
            secondary_y=True
        )

        # Обновление видимых графиков для данного параметра
        visible[i * 2] = True
        visible[i * 2 + 1] = False if i > 0 else True

        # # Добавление кнопки для выбора параметра
        # buttons.append(
        #     dict(
        #         label=parameter,
        #         method="update",
        #         args=[{"visible": visible}, {"title": "График изменения параметра {} по округу {}".format(parameter, adm_area)}]
        #     )
        # )

    # Добавление выпадающего меню для выбора параметра
    fig.update_layout(
        updatemenus=[dict(
            active=0,
            buttons=buttons,
            direction="down",
            showactive=True
        )],
        title="График изменения параметров по округу {}".format(adm_area),
        xaxis_title="Период",
        yaxis_title="MonthlyAverage"
    )
    fig.update_yaxes(title_text="MonthlyAveragePDKss", secondary_y=True)
    fig.show()
draw_graph("Южный административный округ")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df2=df_aggregated[df_aggregated['Period']==max(df_aggregated['Period'])]
# срез данных по параметру и округу
df_slice = df2.loc[(df2['Parameter'].isin(['Оксид углерода', 'Диоксид азота', 'Взвешенные частицы РМ10',  'Взвешенные частицы РМ2.5'])) & (df2['AdmArea'].isin(['Юго-Западный административный округ', 'Западный административный округ', 'Южный административный округ', 'Юго-Восточный административный округ', 'Северо-Западный административный округ']))]

# создание сводной таблицы
piv_table = df_slice.pivot_table(values=['MonthlyAverage','MonthlyAveragePDKss'], index='Parameter', columns='AdmArea', aggfunc='mean')

# отображение первого графика
piv_table['MonthlyAverage'].plot(kind='bar')
plt.title('MonthlyAverage значения параметров по округам')
plt.xlabel('Параметры')
plt.ylabel('Месячное среднее значение')
plt.show()

# отображение второго графика
piv_table['MonthlyAveragePDKss'].plot(kind='bar')
plt.title('MonthlyAveragePDKss значения параметров по округам')
plt.xlabel('Параметры')
plt.ylabel('Месячное среднее значение')
plt.show()
```
